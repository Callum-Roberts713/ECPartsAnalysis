```python
import pandas as pd
import numpy as np
from matplotlib import *
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
from operator import attrgetter
```

```python
UKSanctionsList = pd.read_csv("UKSanctionsList_09-10-24.csv")
UKSanctionsList['Date Designated'] = pd.to_datetime(UKSanctionsList['Date Designated'], format="mixed" , dayfirst=True)

attr_list = ['day', 'month', 'year']
attrs = attrgetter(*attr_list)
UKSanctionsList[attr_list] = UKSanctionsList['Date Designated'].apply(attrs).apply(pd.Series)
```

```python


def PercentagePrint(ColumnName):
    RegimeValues = UKSanctionsList[f"{ColumnName}"].value_counts().to_frame()
    Count = UKSanctionsList[f"{ColumnName}"].count()
    RegimeValues["Percentage of Total"] = (
        RegimeValues["count"].astype(float) / Count
    ) * 100
    print(RegimeValues)
    plt.figure(1)
    plt.style.use("dark_background")
    (RegimeValues["Percentage of Total"]).plot.bar(RegimeValues.index.to_list())
    plt.legend()
    plt.show()
```

```python
PercentagePrint(ColumnName="Regime Name")
```

```python
PercentagePrint(ColumnName="Type")
```

```python
PercentagePrint(ColumnName="Designation source")
```

```python
PercentagePrint(ColumnName="Sanctions Imposed")
```

```python
PercentagePrint(ColumnName="year")
```
